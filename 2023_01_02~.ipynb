#boxplot

```python
df_salary.boxplot(column="salary", return_type="both")
#박스플롯
#return_type="both" 을 하면 다 지정해주는 정보
#없어도 나옴옴 
#o표시들이 이상치치
```

```python
df_salary["salary"].hist()
#이상치 그림으로로
```

```python
q1 = df_salary["salary"].quantile(q=0.25) #25%
q3 = df_salary["salary"].quantile(q=0.75) #75%
IQR = q3-q1
IQR
#salary 는 변수수
#박스 값 생성
#0~25= 1분위 25~50 = 2분위 50~75 = 3분위 75~100 4분위
#박스 값인 2~3분위를 생성성
```

```python
data_IQR = df_salary[(df_salary["salary"]< q3 + IQR * 1.5) & (df_salary["salary"]> q1 - IQR * 1.5)]
#박스
#IQR = 가운데 상자자 
#1.5는 약속
#IQR * 1.5를 더하고 빼면서  상자 위 아래 중간 선 밖에 값을 뺀 데이터를 IQR박스에 넣어버림림
```

```python
data_IQR
```

```python
data_IQR.boxplot(column="salary")
```

```python
data_IQR["salary"].hist()
```

```python
data_outlier = df_salary[(df_salary["salary"]> q3 + IQR *1.5)]
data_outlier
#Q3 + IQR*1.5 = 상자 위 선보다 위에 있는 이상치 값들 = UIF보다 큰값
#Q1 - IQR*15 = LIF보다 작은값값
```

```python
data_outlier["salary"].hist()
```
